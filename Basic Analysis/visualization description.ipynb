```python
import pandas
from google.colab import drive
import pandas
import matplotlib.pyplot as plt
drive.mount('/content/drive')
points = pandas.read_csv('/content/drive/My Drive/EECS4412/dataset/wc_champions.csv')


x = points['champion'].values
y1 = points['sum_total'].values
y2 = points['win_total'].values
y3 = points['sum_blue_side'].values
y4 = points['win_blue_side'].values

df = pandas.DataFrame({'Champion':x, 'sum_total':y1, 'win_total':y2, 'sum_blue_side':y3, 'win_blue_side':y4})

df[["sum_total", "win_total","Champion"]].plot(x='Champion', rot=0,figsize=(30,15),color = ["C2","C3"],kind="bar")

plt.xticks(rotation=90)
plt.show()

df[["sum_blue_side", "win_blue_side","Champion"]].plot(x='Champion', rot=0,figsize=(30,15),color = ["C2","C3"],kind="bar")

plt.xticks(rotation=90)
plt.show()

```

```python
def draw_pie(column,title):
  data_matches = pandas.read_csv('/content/drive/My Drive/EECS4412/dataset/wc_matches.csv')
  data = data_matches[column].value_counts().to_frame().reset_index()
  data.columns = ['name', 'count']
  fig, ax = plt.subplots(figsize=(20, 7), subplot_kw=dict(aspect='equal'))

  wedges, texts = plt.pie(data['count'], labels=data['name'])

  
  ax.set_title(title, pad=20)
  ax.legend(wedges, data['name'],
            title='Teams',
            loc='center left',
            bbox_to_anchor=(1, 0, 0.5, 1))
  plt.axis('equal')
  plt.show()
```

```python
draw_pie('winner','Distribution of the winning percentage by team')
draw_pie('pbp_caster', 'Distribution of the pbp_caster percentage ')
draw_pie('blue', 'Distribution of the blue side percentage by team')
draw_pie('red', 'Distribution of the red side percentage by team')
draw_pie('color_caster', 'Distribution of the color_caster percentage')
```

```python
def draw_bar(column,title):
  data_players = pandas.read_csv('/content/drive/My Drive/EECS4412/dataset/wc_players.csv')
  
  data_players[column].value_counts().plot(kind='bar',figsize=(20,15));
  plt.xlabel(column)
  plt.ylabel('Occurrence')
  plt.title(title)
  plt.show()
draw_bar('champion','Champion Occurrence')
draw_bar('team','Team Occurrence')
draw_bar('ban1','ban1 Occurrence')
draw_bar('ban2','ban2 Occurrence')
draw_bar('ban3','ban3 Occurrence')
```
